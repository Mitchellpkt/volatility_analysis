```python
days_propose_to_pay = 120 # days
span_to_plot = 365*9 # days
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 22})
```

```python
df = pd.read_csv('xmr.csv')
df.head()
```

```python
price_buffer = list()
for r in range(len(df)):
    raw = df.iloc[r].PriceUSD
    if np.isnan(raw):
        print('nan')
        price_buffer.append(np.nan)
    else:
        print('nonnan')
        price_buffer.append(int(round(raw)))
    
df['exchange_rate'] = price_buffer
```

```python
df.head()
```

```python
plt.scatter(range(len(df['date'])), df['exchange_rate'])
plt.scatter(range(span_to_plot), df["exchange_rate"][(len(df)-span_to_plot):len(df)])

```

```python
span_to_plot = 9*365
plt.scatter(range(len(df['date'])), df['exchange_rate'], color='black')
plt.scatter(range((len(df)-2*span_to_plot),len(df)), df["exchange_rate"][(len(df)-2*span_to_plot):len(df)], color='green')
plt.scatter(range((len(df)-span_to_plot),len(df)), df["exchange_rate"][(len(df)-span_to_plot):len(df)], color='blue')
plt.xlabel('Days since April 2014')
plt.ylabel('USD/XMR')
```

```python
df.to_csv('xmr_exchange_rate_all.csv', columns=('date','exchange_rate'), index=False)
```

```python
ROI_buffer = list(np.full(days_propose_to_pay, np.nan))
for r in range(days_propose_to_pay, len(df)):
    price_propose = df.iloc[r-days_propose_to_pay].PriceUSD
    price_payout = df.iloc[r].PriceUSD
    # print('*****')
    # print('Price at proposal: ' + str(price_propose))
    # print('Price at payout: ' + str(price_payout))
    ROI_buffer.append(100*price_payout/price_propose-100)
    
df['propose_to_pay'] = ROI_buffer
```

```python
plt.scatter(range(len(df)), df['propose_to_pay'])
```

```python
f = plt.figure(figsize=(10,5), facecolor='white')
plt.hist(df["propose_to_pay"][(len(df)-span_to_plot):len(df)],bins=200)
plt.xlabel("Percent increase or decrease w.r.t. USD")
plt.ylabel('Count (days)')
plt.title('Change in XMR/USD over last 4 years');
plt.axvline(x=0,color='red')
plt.xlim((250,2000))
plt.ylim((0,15))
```

```python
f = plt.figure(figsize=(10,5), facecolor='white')
plt.hist(df["propose_to_pay"][(len(df)-span_to_plot):len(df)],bins=200, cumulative=True, density=True)
plt.xlabel("Percent increase or decrease in value (wrt USD)");
plt.ylabel("Norm'd cumulative count");
plt.title('120 day volatility over last 4 years')
plt.xlim((min(df['propose_to_pay'].dropna()),max(df['propose_to_pay'].dropna())));
plt.axhline(y=0.4, color='red',linewidth=3)
plt.axvline(x=0, color='red',linewidth=3)
plt.axhline(y=0.8, color='orange',linewidth=3)
plt.axvline(x=170, color='orange',linewidth=3)
```
